```python
# Image Denoising using Classical Filters
## Gaussian vs Median vs Proposed (Median + Bilateral)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("D:\\imgNoise.jpg")

if img is None:
    raise ValueError("(imgNoise.jpg)")

if len(img.shape) == 3:
    noisy = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
else:
    noisy = img.copy()

gaussian = cv2.GaussianBlur(noisy, (5, 5), 1.0)

median = cv2.medianBlur(noisy, 5)

step1 = median
final = cv2.bilateralFilter(step1, d=9, sigmaColor=75, sigmaSpace=75)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(noisy, cmap='gray')
plt.title("Noisy Image")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(gaussian, cmap='gray')
plt.title("Gaussian Filter")
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(median, cmap='gray')
plt.title("Median Filter (k=5)")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(final, cmap='gray')
plt.title("Proposed: Median + Bilateral")
plt.axis('off')

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(noisy, cmap='gray')
plt.title("Noisy")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(final, cmap='gray')
plt.title("Denoised (Proposed)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
